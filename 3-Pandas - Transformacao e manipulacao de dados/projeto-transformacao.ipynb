```python
dados_projecto_1 = panda.read_json('./dados_vendas_clientes.json')
dados_projecto_1 = panda.json_normalize(dados_projecto_1['dados_vendas'])
dados_projecto_1
```

__Aula 2__

```python
colunas_dados_projecto_1 = list(dados_projecto_1.columns)
colunas_dados_projecto_1
```

```python
dados_projecto_1.head()
```

```python
dados_projecto_1 = dados_projecto_1.explode(colunas_dados_projecto_1[1:])
dados_projecto_1.reset_index(inplace=True, drop=True)
dados_projecto_1
```

```python
dados_projecto_1.info()
```

```python
dados_projecto_1['Valor da compra'] = dados_projecto_1['Valor da compra'].apply(lambda x: x.replace('R$ ','').replace(',','.'))
dados_projecto_1['Valor da compra'] = dados_projecto_1['Valor da compra'].astype(numpy.float64)
dados_projecto_1
```

```python
dados_projecto_1.info()
```

```python
### Projeto desafio 1: vendas online


# Transformar os textos de Cliente para texto em minúsculo
dados_projecto_1['Cliente'] = dados_projecto_1['Cliente'].str.lower()
# Verificar o resultado
dados_projecto_1.head()

# Opção de substituição - necessário verificar o resultado da substituição
# O regex não seleciona todas as letras de a-z e espaços em branco ' '
# Tudo que satisfaz o regex é apagado
dados_projecto_1['Cliente'].str.replace('[^a-z ]', '', regex=True)

# Realizar a substituição dos dados na coluna textual
dados_projecto_1['Cliente'] = dados_projecto_1['Cliente'].str.replace('[^a-z ]', '', regex=True).str.strip()
# Visualizar o resultado final
dados_projecto_1.head()
```

Aula 4

```python
# Transformar para o tipo datetime definindo o formato de data como DD/MM/AAAA ('%d/%m/%Y')
dados_projecto_1['Data de venda'] = panda.to_datetime(dados_projecto_1['Data de venda'], format='%d/%m/%Y')
# Visualizar o resultado
dados_projecto_1

# Verificar o resultado da transformação
dados_projecto_1.info()

# Calcular o total arrecadado em compras por cada cliente
total_compras = dados_projecto_1.groupby(['Cliente'])['Valor da compra'].sum()
# Visualizar o resultado
total_compras
```

**ETAPAS - PROJETO 2**

```python
dados_projecto_2 = panda.read_json('./dados_locacao_imoveis.json')
dados_projecto_2 = panda.json_normalize(dados_projecto_2['dados_locacao'])
dados_projecto_2
```

__Aula 2__

```python
colunas_dados_projecto_2 = list(dados_projecto_2.columns)
colunas_dados_projecto_2
```

```python
dados_projecto_2 = dados_projecto_2.explode(colunas_dados_projecto_2[1:])
dados_projecto_2.reset_index(inplace=True, drop=True)
dados_projecto_2
```

```python
dados_projecto_2.info()
```

```python
dados_projecto_2['valor_aluguel'] = dados_projecto_2['valor_aluguel'].apply(lambda x: x.replace('$ ','').replace(',','.').replace(' reais',''))
dados_projecto_2['valor_aluguel'] = dados_projecto_2['valor_aluguel'].astype(numpy.float64)
dados_projecto_2
```

```python
dados_projecto_2.info()
```

```python
# Opção de substituição - necessário verificar o resultado da substituição
# Foi necessário adicionar a barra '\' para ser considerados os parênteses como caracteres literais
dados_projecto_2['apartamento'].str.replace(' \(blocoAP\)', '', regex=True)

# Realizar a substituição dos dados na coluna textual
dados_projecto_2['apartamento'] = dados_projecto_2['apartamento'].str.replace(' \(blocoAP\)', '', regex=True)
# Visualizar o resultado final
dados_projecto_2
```

Aula 4

```python
# Transformar para o tipo datetime definindo o formato de data como DD/MM/AAAA ('%d/%m/%Y')
dados_projecto_2['datas_combinadas_pagamento'] = panda.to_datetime(dados_projecto_2['datas_combinadas_pagamento'], format='%d/%m/%Y')
dados_projecto_2['datas_de_pagamento'] = panda.to_datetime(dados_projecto_2['datas_de_pagamento'], format='%d/%m/%Y')

# Visualizar o resultado
dados_projecto_2

# Para contribuir na solução do contexto é possível calcular a diferença de dias
# entre a data combinada e a data de pagamento com dt.days
dados_projecto_2['atraso'] = (dados_projecto_2['datas_de_pagamento'] - dados_projecto_2['datas_combinadas_pagamento']).dt.days

# Observar o novo DataFrame
dados_projecto_2

# Calcular a média de tempo de atraso por apartamentos
media_atraso = dados_projecto_2.groupby(['apartamento'])['atraso'].mean()

# Visualizar o resultado
media_atraso
```
